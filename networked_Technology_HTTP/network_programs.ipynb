```python
import socket
i = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
i.connect(('data.pr4e.org', 80))
cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
i.send(cmd)

while True :
    data = i.recv(100)
    if(len(data)< 1):
        break
    print(data.decode())
i.close()
```
